```sql
CREATE OR REPLACE FUNCTION hours_above_temperature_plant_schedule_udtf (x int)
RETURNS TABLE(zone string, date_format date, hours_above_temp int, percent_plant_on double)
COMMENT "Addresses questions related to how many hours does the temperature exceed x degrees across the zones and how does it correspond to the % of hours the plant is turned on?"
RETURN
with above_temp_plant_on_flag 
  (SELECT 'SPACE1-1' AS Space,
  beatrice_liew.arup.parse_date(`Date/Time`) as date_format,
  CASE
      WHEN `SPACE1-1:Zone Air Temperature [C](Hourly)` > x THEN 1
      ELSE 0
    END AS Hours_above_temp,
      CASE
        WHEN `SPACE1-1:Zone Air Temperature [C](Hourly)` > x
        AND `PLANTONOFFSCHEDULE:Schedule Value [](Hourly)` = 1 THEN 1
        ELSE 0
      END as hours_above_temp_with_plant_on
FROM
  beatrice_liew.arup.engineering_results_raw
UNION ALL
SELECT
  'SPACE2-1' AS Space,
  beatrice_liew.arup.parse_date(`Date/Time`) as date_format,
  CASE
      WHEN `SPACE2-1:Zone Air Temperature [C](Hourly)` > x THEN 1
      ELSE 0
    END AS Hours_above_temp,
      CASE
        WHEN `SPACE2-1:Zone Air Temperature [C](Hourly)` > x
        AND `PLANTONOFFSCHEDULE:Schedule Value [](Hourly)` = 1 THEN 1
        ELSE 0
      END as hours_above_temp_with_plant_on
FROM
  beatrice_liew.arup.engineering_results_raw
UNION ALL
SELECT
  'SPACE3-1' AS Space,
  beatrice_liew.arup.parse_date(`Date/Time`) as date_format,
  CASE
      WHEN `SPACE3-1:Zone Air Temperature [C](Hourly)` > x THEN 1
      ELSE 0
    END AS Hours_above_temp,
      CASE
        WHEN `SPACE3-1:Zone Air Temperature [C](Hourly)` > x
        AND `PLANTONOFFSCHEDULE:Schedule Value [](Hourly)` = 1 THEN 1
        ELSE 0
      END as hours_above_temp_with_plant_on
FROM
  beatrice_liew.arup.engineering_results_raw
UNION ALL
SELECT
  'SPACE4-1' AS Space,
  beatrice_liew.arup.parse_date(`Date/Time`) as date_format,
  CASE
      WHEN `SPACE4-1:Zone Air Temperature [C](Hourly)` > x THEN 1
      ELSE 0
    END AS Hours_above_temp,
      CASE
        WHEN `SPACE4-1:Zone Air Temperature [C](Hourly)` > x
        AND `PLANTONOFFSCHEDULE:Schedule Value [](Hourly)` = 1 THEN 1
        ELSE 0
      END as hours_above_temp_with_plant_on
FROM
  beatrice_liew.arup.engineering_results_raw
UNION ALL
SELECT
  'SPACE5-1' AS Space,
  beatrice_liew.arup.parse_date(`Date/Time`) as date_format,
  CASE
      WHEN `SPACE5-1:Zone Air Temperature [C](Hourly)` > x THEN 1
      ELSE 0
    END AS Hours_above_temp,
      CASE
        WHEN `SPACE5-1:Zone Air Temperature [C](Hourly)` > x
        AND `PLANTONOFFSCHEDULE:Schedule Value [](Hourly)` = 1 THEN 1
        ELSE 0
      END as hours_above_temp_with_plant_on
FROM
  beatrice_liew.arup.engineering_results_raw
  )

select Space, date_format, sum(Hours_above_temp) as hours_above_temperature, 
(sum(hours_above_temp_with_plant_on) / sum (Hours_above_temp) * 100) percent_plant_on
from above_temp_plant_on_flag
group by Space, date_format
```
