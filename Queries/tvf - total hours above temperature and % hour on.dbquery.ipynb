```sql
CREATE OR REPLACE FUNCTION hours_below_temperature_plant_schedule_udtf (x int)
RETURNS TABLE(zone string, hours_exceed_24 int, Percent_Hours_Plant_On double)
COMMENT "How many hours does the temperatures in the zones exceed x degrees and what is the corresponding % hours against the plant is on??"
RETURN
SELECT
  'SPACE1-1' AS Zone,
  SUM(
    CASE
      WHEN `SPACE1-1:Zone Air Temperature [C](Hourly)` > x THEN 1
      ELSE 0
    END
  ) AS Hours_Exceed_24,
  (
    SUM(
      CASE
        WHEN `SPACE1-1:Zone Air Temperature [C](Hourly)` > x
        AND `PLANTONOFFSCHEDULE:Schedule Value [](Hourly)` = 1 THEN 1
        ELSE 0
      END
    ) / SUM(
      CASE
        WHEN `SPACE1-1:Zone Air Temperature [C](Hourly)` > x THEN 1
        ELSE 0
      END
    )
  ) * 100 AS Percent_Hours_Plant_On
FROM
  beatrice_liew.arup.engineering_results_raw
UNION ALL
SELECT
  'SPACE2-1' AS Zone,
  SUM(
    CASE
      WHEN `SPACE2-1:Zone Air Temperature [C](Hourly)` > x THEN 1
      ELSE 0
    END
  ) AS Hours_Exceed_24,
  (
    SUM(
      CASE
        WHEN `SPACE2-1:Zone Air Temperature [C](Hourly)` > x
        AND `PLANTONOFFSCHEDULE:Schedule Value [](Hourly)` = 1 THEN 1
        ELSE 0
      END
    ) / SUM(
      CASE
        WHEN `SPACE2-1:Zone Air Temperature [C](Hourly)` > x THEN 1
        ELSE 0
      END
    )
  ) * 100 AS Percent_Hours_Plant_On
FROM
  beatrice_liew.arup.engineering_results_raw
UNION ALL
SELECT
  'SPACE3-1' AS Zone,
  SUM(
    CASE
      WHEN `SPACE3-1:Zone Air Temperature [C](Hourly)` > x THEN 1
      ELSE 0
    END
  ) AS Hours_Exceed_24,
  (
    SUM(
      CASE
        WHEN `SPACE3-1:Zone Air Temperature [C](Hourly)` > x
        AND `PLANTONOFFSCHEDULE:Schedule Value [](Hourly)` = 1 THEN 1
        ELSE 0
      END
    ) / SUM(
      CASE
        WHEN `SPACE3-1:Zone Air Temperature [C](Hourly)` > x THEN 1
        ELSE 0
      END
    )
  ) * 100 AS Percent_Hours_Plant_On
FROM
  beatrice_liew.arup.engineering_results_raw
UNION ALL
SELECT
  'SPACE4-1' AS Zone,
  SUM(
    CASE
      WHEN `SPACE4-1:Zone Air Temperature [C](Hourly)` > x THEN 1
      ELSE 0
    END
  ) AS Hours_Exceed_24,
  (
    SUM(
      CASE
        WHEN `SPACE4-1:Zone Air Temperature [C](Hourly)` > x
        AND `PLANTONOFFSCHEDULE:Schedule Value [](Hourly)` = 1 THEN 1
        ELSE 0
      END
    ) / SUM(
      CASE
        WHEN `SPACE4-1:Zone Air Temperature [C](Hourly)` > x THEN 1
        ELSE 0
      END
    )
  ) * 100 AS Percent_Hours_Plant_On
FROM
  beatrice_liew.arup.engineering_results_raw
UNION ALL
SELECT
  'SPACE5-1' AS Zone,
  SUM(
    CASE
      WHEN `SPACE5-1:Zone Air Temperature [C](Hourly)` > x THEN 1
      ELSE 0
    END
  ) AS Hours_Exceed_24,
  (
    SUM(
      CASE
        WHEN `SPACE5-1:Zone Air Temperature [C](Hourly)` > x
        AND `PLANTONOFFSCHEDULE:Schedule Value [](Hourly)` = 1 THEN 1
        ELSE 0
      END
    ) / SUM(
      CASE
        WHEN `SPACE5-1:Zone Air Temperature [C](Hourly)` > x THEN 1
        ELSE 0
      END
    )
  ) * 100 AS Percent_Hours_Plant_On
FROM
  beatrice_liew.arup.engineering_results_raw
```
